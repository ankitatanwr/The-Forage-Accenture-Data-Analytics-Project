```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:/Users/intel/Desktop/Reactions_Cleaned_Data.xlsx"
sheets = pd.read_excel(file_path, sheet_name=None)
cleaned_data = sheets['Cleaned Data']
aggregate_score = sheets['Aggregrate Score']

cleaned_data['DateTime'] = pd.to_datetime(cleaned_data['DateTime'])

plt.figure(figsize=(10, 6))
plt.bar(aggregate_score['Category'], aggregate_score['Aggregrate score'], color='blue')
plt.title('Aggregate Scores for Each Category')
plt.xlabel('Category')
plt.ylabel('Aggregate Score')
plt.xticks(rotation=45)
plt.show()
```

```python
cleaned_data['DateTime'] = pd.to_datetime(cleaned_data['DateTime'])  # Ensure DateTime is in datetime format
time_series_data = cleaned_data.groupby(cleaned_data['DateTime'].dt.date)['Reaction Type'].count()

plt.figure(figsize=(12, 6))
time_series_data.plot(kind='line', color='purple')
plt.title('Number of Reactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reactions')
plt.grid(True)
plt.show()
```

```python
sheets = pd.read_excel(file_path, sheet_name=None)
cleaned_data = sheets['Cleaned Data']

reaction_counts_cleaned = cleaned_data['Reaction Type'].value_counts()

plt.figure(figsize=(10, 6))
reaction_counts_cleaned.plot(kind='bar', color='skyblue')
plt.title('Count of Each Reaction Type (Cleaned Data)')
plt.xlabel('Reaction Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
categories = aggregate_score['Category']
scores = aggregate_score['Aggregrate score']
plt.figure(figsize=(8, 8))
plt.pie(scores, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Aggregate Score by Category')
plt.show()
```
